```python
from ultralytics import YOLO
```

```python
!unzip -q "/content/ForesFireDataset(ObjectDetection).zip" -d /content/fire_detect
```

```python
import os
os.listdir("/content/fire_detect")
```

```python
%%writefile fire.yaml
path: /content/fire_detect
train: train/images
val: valid/images
test: test/images

names:
  0: fire
```

```python
model = YOLO("yolov8n.pt")
```

```python
model.train(
    data="fire.yaml",
    epochs=25,
    imgsz=640,
    batch=16
)
```

```python
model.predict("/content/fire_detect/test/images", save=True)
```

```python
model.export(format="onnx")
```
